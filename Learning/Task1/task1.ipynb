```python
from docx import Document

def count_docx_stats(file_path):
    doc = Document(file_path)

    # Counting pages based on '\f' character
    num_pages = len(doc.sections)  # Assuming each section represents a page

    # Counting lines based on '\n' character
    num_lines = sum(len(p.runs) for p in doc.paragraphs)

    # Counting paragraphs based on paragraphs
    num_paragraphs = len(doc.paragraphs)

    # Counting words based on words
    num_words = sum(len(p.text.split()) for p in doc.paragraphs)

    return num_pages, num_lines, num_paragraphs, num_words

# Example usage:
file_path = 'sample_text.docx'  # Update with your file path
pages, lines, paragraphs, words = count_docx_stats(file_path)
print("Pages:", pages)
print("Lines:", lines)
print("Paragraphs:", paragraphs)
print("Words:", words)
```

```python
"""String Functions in Python """
# Sample string for demonstration
sample_string = "Python is a versatile programming language."

# 1. String Length
print("Length:", len(sample_string))

# 2. Upper and Lower Case
print("Upper Case:", sample_string.upper())
print("Lower Case:", sample_string.lower())

# 3. Capitalize the First Letter
print("Capitalized:", sample_string.capitalize())

# 4. Count occurrences of a substring
print("Occurrences of 'a':", sample_string.count('a'))

# 5. Find the index of a substring
print("Index of 'versatile':", sample_string.find('versatile'))

# 6. Replace a substring
print("Replace 'versatile' with 'flexible':", sample_string.replace('versatile', 'flexible'))

# 7. Split a string into a list based on a delimiter
print("Split:", sample_string.split())

# 8. Join elements of a list into a string
words = ['Python', 'is', 'a', 'versatile', 'programming', 'language']
print("Join:", ' '.join(words))

# 9. Check if the string starts or ends with a substring
print("Starts with 'Python':", sample_string.startswith('Python'))
print("Ends with 'language':", sample_string.endswith('language'))

# 10. Strip leading and trailing whitespace
string_with_whitespace = "   Python   "
print("Stripped:", string_with_whitespace.strip())

# 11. Check if the string contains only alphabetic characters
print("Is Alpha:", sample_string.isalpha())

# 12. Check if the string contains only numeric characters
numeric_string = "12345"
print("Is Numeric:", numeric_string.isnumeric())

# 13. Check if the string contains only alphanumeric characters
alphanumeric_string = "Python123"
print("Is Alphanumeric:", alphanumeric_string.isalnum())

# 14. Check if the string contains only whitespace
whitespace_string = "   "
print("Is Whitespace:", whitespace_string.isspace())
```
