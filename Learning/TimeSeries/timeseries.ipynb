```python
import pandas as pd
import numpy as np

data = pd.read_csv("nifty3monthdata.csv")
data.head()
data.info()
data.describe()
```

```python
missing_values = data.isnull().sum()
missing_values
```

```python
data.head()
```

```python
import matplotlib.pyplot as plt
data['Date '] = pd.to_datetime(data['Date '])

# Plotting the data
fig, axs = plt.subplots(6, 1, figsize=(12, 18))

columns = ["Open ", "High ", "Low ", "Close ", "Shares Traded ", "Turnover (₹ Cr)"]
for i, col in enumerate(columns):
    axs[i].plot(data['Date '], data[col], marker='o')
    axs[i].set_title(col)
    axs[i].grid(True)

# Formatting x-axis with date labels
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()

```

```python
# Ensure there are no leading or trailing spaces in column names
data.columns = data.columns.str.strip()

# Converting 'Date' column to datetime
data['Date'] = pd.to_datetime(data['Date'], format='%d-%b-%Y')

# Plotting 'Open' and 'Close' prices against 'Date'
plt.figure(figsize=(12, 6))
plt.plot(data['Date'], data['Open'], marker='o', label='Open')
plt.plot(data['Date'], data['Close'], marker='o', label='Close')
plt.title('Open and Close Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.grid(True)
plt.legend()

# Formatting x-axis with date labels
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```
